1. Create a line chart that plots the sample mean as more and more trials are conducted, up to 1,000 trials.

2. Randomly sample the mean 1,000 times, and plot the resulting frequency of sample means as a histogram.


```
# Initialize environment
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from numpy.random import randint
from numpy import mean
```

```
# normal distribution

#    create variable list for results
results = []

for n_trials in range(1000):
    trials = np.random.normal(3, 1, n_trials)
    m_trials = trials.mean()
    results.append(m_trials)

df = pd.DataFrame({'trials' : results})

df.plot()
```

```
s_means = [mean(np.random.normal(3, 1, 1000)) for _ in range(3000)]

plt.hist(s_means)
plt.show()
```

```
# binomial distribution

#    create variable list for results
results = []

for n_trials in range(1000):
    trials = np.random.binomial(5, np.random.beta(200, 200), n_trials)
    m_trials = trials.mean()
    results.append(m_trials)

df = pd.DataFrame({'trials' : results})

df.plot()
```

```
means = [mean(np.random.binomial(5, np.random.beta(200,200), n_trials)) for _ in range(1000)]

plt.hist(means)
plt.show()
```
